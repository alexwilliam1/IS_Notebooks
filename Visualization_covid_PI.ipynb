## üìö Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt 
```

## üé≤ Carregando os dados

```python
DATASET_PATH = 'C://Users//alexw//Documents//UFPI//Sistemas_Inteligentes//04-01-21_visualization_pre_processing//'
DATASET_NAME = 'covid-piaui.csv'
```

```python
def load_data(dataset_path, dataset_name):
    csv_path = os.path.join(dataset_path, dataset_name)
    return pd.read_csv(csv_path)
```

```python
covid = load_data(DATASET_PATH,DATASET_NAME)
```

## üìã Verificando a estrutura dos dados

```python
covid = covid.drop(["city_ibge_code", "is_last"], axis=1)
covid
```

## üìä Histogramas dos dados n√£o categ√≥ricos

```python
covid.hist(bins=75, figsize=(20,15))
plt.show()
```

## üåç Gr√°fico de Latitude e Longitude

```python
multi = 2
covid.plot(kind='scatter', x='longitude', y='latitude', alpha=0.3,
    s=covid['estimated_population']/100, label='estimated_population', figsize=(10*multi,7*multi),
    c='death_rate', cmap=plt.get_cmap('jet'), colorbar=True,
    sharex=False)
plt.show()
```

## ‚ìÇÔ∏è Matriz de correla√ß√£o
